```python
# Definicja klasy
class Counter:
    def __init__(self, value = 0):
        self.count = value

    def add(self, value):
        self.count += value

    def subtract(self, value):
        self.count -= value

    def zero(self):
        self.count = 0

    def get_value(self):
        return self.count

# Przykładowe użycie
licznik1 = Counter()
print(licznik1.get_value())
licznik1.add(3)
print(licznik1.get_value())
licznik1.subtract(1)
print(licznik1.get_value())

licznik2 = Counter(10)
print(licznik2.get_value())
```

### ZADANIE 2

```python
# Tworzenie obiektu dataclass
from dataclasses import dataclass

@dataclass
class Measurement:
    time: str
    value: float
    unit: str

# Przykładowe użycie dataclass
pomiar1 = Measurement(time='11:11', value=40.0, unit='g')
pomiar2 = Measurement(time='11:45', value=50.0, unit='g')
pomiar3 = Measurement(time='12:02', value=90.0, unit='g')
pomiar4 = Measurement(time='12:30', value=80.0, unit='g')

pomiary = [pomiar1, pomiar2, pomiar3, pomiar4]

for pomiar in pomiary:
    print(f'Czas pomiaru: {pomiar.time}, Wartość: {pomiar.value} {pomiar.unit}')

for pomiar in pomiary:
    max_value = 0
    max_measurement = None
    if pomiar.value > max_value:
        max_value = pomiar.value
        max_measurement = pomiar
print(f'Pomiar o największej wartości: {max_measurement.time} - {max_measurement.value} {max_measurement.unit}')
```

### ZADANIE 3

```python
# Tworzenie obiektu
class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def print(self):
        print(f'[{self.x}, {self.y}]')

    def __add__(self, other):
        return Vector2D(self.x + other.x, self.y + other.y)

    def __mul__(self, scalar):
        return Vector2D(self.x * scalar, self.y * scalar)

    def __rmul__(self, scalar):
        return Vector2D(self.x * scalar, self.y * scalar)

    # Mnożenie skalarne dwóch wektorów
    def multiply(self, other):
        return self.x * other.x + self.y * other.y

# Przykładowe użycie
wektor1 = Vector2D(1, 2)
wektor1.print()

wektor2 = Vector2D(3, 4)
wektor2.print()

wektor3 = wektor1 + wektor2
wektor3.print()

wektor4 = wektor1 * 2
wektor4.print()

wektor5 = 2 * wektor1
wektor5.print()

skalar = wektor1.multiply(wektor2)
print(skalar)

```

```python
# Tworzenie pojemnika na pomiary
class MeasurementSeries:
    def __init__(self, measurements):
        self.measurements = measurements
    
    def __getitem__(self, index):
        return self.measurements[index]

    def __setitem__(self, index, value):
        self.measurements[index] = value

    def __len__(self):
        return len(self.measurements)

    def __iter__(self):
        return iter(self.measurements)

    def max_value(self):
        max_measurement = max(self.measurements, key=lambda m: m.value)
        return max_measurement

    def min_value(self):
        min_measurement = min(self.measurements, key=lambda m: m.value)
        return min_measurement

    def mean(self):
        total = 0
        for measurement in self.measurements:
            total += measurement.value
        return total / len(self.measurements)

# Przykładowe użycie
pomiary = MeasurementSeries([pomiar1, pomiar2, pomiar3, pomiar4])
print(f'Pierwszy pomiar: {pomiary[0]}')

print(f'Drugi pomiar przed zmianą: {pomiary[1]}')
pomiar5 = Measurement(time='13:00', value=200.0, unit='g')
pomiary[1] = pomiar5
print(f'Drugi pomiar po zmianie: {pomiary[1]}')

print(f'Liczba pomiarów: {len(pomiary)}')

print('Wszystkie pomiary:')
for pomiar in pomiary:
    print(pomiar)
    
print(f'Pomiar o największej wartości: {pomiary.max_value()}')
print(f'Pomiar o najmniejszej wartości: {pomiary.min_value()}')
print(f'Średnia wartość pomiarów: {pomiary.mean()}{pomiary[0].unit}')
```

```python
# Tworezenie obiektu
class Polynomial:
    def __init__(self, coefficients):
        self.coefficients = coefficients
        # Usuwanie zerowych współczynników z końca listy
        while len(self.coefficients) > 1 and self.coefficients[-1] == 0:
            self.coefficients.pop()

    def print(self):
        terms = []
        power = 0
        for coeffcient in self.coefficients:
            if power == 0:
                terms.append(f'{coeffcient}')
            elif power == 1:
                terms.append(f'{coeffcient}x')
            else:
                terms.append(f'{coeffcient}x^{power}')
            power += 1
        print(" + ".join(terms))

    def __call__(self, x):
        result = 0
        power = 0
        for coeffcient in self.coefficients:
            result += coeffcient * (x ** power)
            power += 1
        return result

    def differentiate(self):
        if len(self.coefficients) == 1:
            return Polynomial([0])
        new_coefficients = []
        power = 1
        for coeffcient in self.coefficients[1:]:
            new_coefficients.append(coeffcient * power)
            power += 1
        return Polynomial(new_coefficients)    

# Przykładowe użycie
wielomian1 = Polynomial([1, 0, 3])
wielomian1.print()

print(wielomian1(2))

pochodna = wielomian1.differentiate()
pochodna.print()
```

### ZADANIE 6

```python
# Tworzenie obiektu do dziedziczenia
class Account:
    def __init__(self, username, points):
        self.username = username
        self.points = points

    def add_points(self, points):
        self.points += points

    def remove_points(self, points):
        self.points -= points

    def info(self):
        print(f'Username: {self.username}, Points: {self.points}')

class PremiumAccount(Account):
    def __init__(self, username, points, premium_multiplier):
        super().__init__(username, points)
        self.premium_multiplier = premium_multiplier

    def add_points(self, points):
        super().add_points(points * self.premium_multiplier)

class PenaltyAccount(Account):
    def __init__(self, username, points, penalty_multiplier):
        super().__init__(username, points)
        self.penalty_multiplier = penalty_multiplier

    def remove_points(self, points):
        super().remove_points(points * self.penalty_multiplier)

# Przykładowe użycie
konto1 = Account('user_basic', 100)
konto1.info()
konto1.add_points(50)
konto1.info()
konto1.remove_points(20)
konto1.info()

konto2 = PremiumAccount('user_premium', 100, 2)
konto2.info()
konto2.add_points(50)
konto2.info()

konto3 = PenaltyAccount('user_penalty', 100, 2)
konto3.info()
konto3.remove_points(20)
konto3.info()
```

### ZADANIE 7

```python
from abc import ABC, abstractmethod

# Tworzenie klasy abstrakcyjnej
class Animal(ABC):
    def __init__(self, name, x, y):
        self.name = name
        self.x = x
        self.y = y

    def info(self):
        print(f'{self.name}: ({self.x}, {self.y})')

    @abstractmethod
    def move(self):
        pass
    
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def __init__(self, name, x, y):
        super().__init__(name, x, y)

    def sound(self):
        print('Woof!')
    
    def move(self):
        self.x += 3
        print(f'{self.name} biegnie na pole ({self.x}, {self.y})')

class Bird(Animal):
    def sound(self):
        print('Tweet!')
    
    def move(self):
        self.y += 4
        print(f'{self.name} lata na pole ({self.x}, {self.y})')

# Przykładowe użycie
pies = Dog('Reksio', 1, 3)
ptak = Bird('Kanar', 2, 1)

pies.info()
pies.move()
pies.sound()

ptak.info()
ptak.move()
ptak.sound()
```
