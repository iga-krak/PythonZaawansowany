```python
import time

class TimerDecorator:
    def __init__(self, func=None, print_each_call=False, max_history=None):
        self.func = func
        self.print_each_call = print_each_call
        self.max_history = max_history
        self.execution_times = []
        
        if self.func is not None:
            self.__name__ = self.func.__name__

    def __call__(self, *args, **kwargs):
        if self.func is None:
            self.func = args[0]
            self.__name__ = self.func.__name__
            return self
            
        start_time = time.perf_counter()
        result = self.func(*args, **kwargs)
        end_time = time.perf_counter()
        exec_time = end_time - start_time
        self.execution_times.append(exec_time)

        if self.max_history is not None:
            if len(self.execution_times) > self.max_history:
                self.execution_times.pop(0)
        
        if self.print_each_call:
            print(f"{self.__name__} execution time: {exec_time:.6f} s")
            
        return result

    def get_stats(self):
        print(f"\nStatystyki funkcji: {self.__name__}")
        calls = len(self.execution_times)
        
        if self.max_history is not None:
            print(f"Liczba zapisanych pomiarów: {calls} [limit: {self.max_history}]")
        else:
            print(f"Liczba zapisanych pomiarów: {calls} [no limit]")
        
        if calls > 0:
            min_time = min(self.execution_times)
            max_time = max(self.execution_times)
            avg_time = sum(self.execution_times) / calls
            
            print(f"Najkrótszy czas: {min_time:.6f} s")
            print(f"Najdłuższy czas: {max_time:.6f} s")
            print(f"Średni czas:     {avg_time:.6f} s")

    def clear_stats(self):
        self.execution_times.clear()
        
# TESTOWANIE FUNKCJI
@TimerDecorator
def obliczenia(n, delay=0.01):
    time.sleep(delay)
    return sum(i * i for i in range(n))

@TimerDecorator(print_each_call=True, max_history=3)
def obliczenia_wypisywanie(n, delay=0.01):
    time.sleep(delay)
    return sum(i * i for i in range(n))

obliczenia(10000)
obliczenia(20000)
obliczenia(30000)
obliczenia.get_stats()

obliczenia_wypisywanie(5, delay=0.02)
obliczenia_wypisywanie(10, delay=0.03)
obliczenia_wypisywanie(15, delay=0.01)
obliczenia_wypisywanie.get_stats()

obliczenia_wypisywanie(20, delay=0.09)
obliczenia_wypisywanie.get_stats()

obliczenia_wypisywanie.clear_stats()
obliczenia_wypisywanie.get_stats()
```

```python
class Temperature:
    # Stała klasowa ułatwiająca czytelność kodu
    absolute_zero_celsius = -273.15

    def __init__(self, degrees_celsius):
        self.celsius = degrees_celsius

    @property
    def celsius(self):
        return round(self._celsius, 2)

    @celsius.setter
    def celsius(self, value):
        if value < self.absolute_zero_celsius:
            self._celsius = self.absolute_zero_celsius  
        else:
            self._celsius = value

    @property
    def fahrenheit(self):
        return round((self.celsius * 9 / 5) + 32, 2)

    @fahrenheit.setter
    def fahrenheit(self, value):
        self.celsius = (value - 32) * 5 / 9

    @property
    def kelvin(self):
        return round(self.celsius - self.absolute_zero_celsius, 2)

    @kelvin.setter
    def kelvin(self, value):
        self.celsius = value + self.absolute_zero_celsius

    def __repr__(self):
        return f"Temperature({self.celsius:.2f}°C)"
    
    @property
    def is_freezing(self):
        return self.celsius <= 0
    
    @property
    def is_boiling(self):
        return self.celsius >= 100
    
    @property
    def state(self):
        if self.celsius < 0:
            return "solid"
        elif self.celsius < 100:
            return "liquid"
        else:
            return "gas"
    

# PRZYKŁADOWE UŻYCIE
t = Temperature(25)

print(t.celsius)      # 25
print(t.fahrenheit)   # 77.0
print(t.kelvin)       # 298.15

t.fahrenheit = 32
print(t.celsius)      # 0.0

t.kelvin = 300
print(t.celsius)      # 26.85

t.kelvin = -300      
print(t.kelvin)      # 0

print (t)

print(t.is_freezing)  # True
print(t.is_boiling)   # False

t.celsius = -10
print(t.state)        # solid

t.celsius = 50
print(t.state)        # liquid

t.celsius = 150
print(t.state)        # gas
```

```python
def lcg(seed, a, c, m, N, alternative_mode=False):
    xn = seed
    
    for _ in range(N):
        xn = (a * xn + c) % m
        if alternative_mode:
            yield xn / m
        else:
            yield xn


# ==========================================
# DEMONSTRACJA DZIAŁANIA
# ==========================================

seed = 1
a = 5
c = 3
m = 16
N = 10

generator_liczb = lcg(seed, a, c, m, N)

x1 = next(generator_liczb)
print(f"x1: {x1}")
    
x2 = next(generator_liczb)
print(f"x2: {x2}")
    
x3 = next(generator_liczb)
print(f"x3: {x3}")

print("\nKolejne liczby z generatora (pętla for):")
for liczba in generator_liczb:
    print(f"xn: {liczba}")

print("\n=== Alternatywny tryb (liczby z zakresu [0, 1)): ===")

generator_liczb = lcg(seed, a, c, m, N, alternative_mode=True)

x1 = next(generator_liczb)
print(f"x1: {x1}")
    
x2 = next(generator_liczb)
print(f"x2: {x2}")
    
x3 = next(generator_liczb)
print(f"x3: {x3}")

print("\nKolejne liczby z generatora (pętla for):")
for liczba in generator_liczb:
    print(f"xn: {liczba}")
```
