```python
class ElectronicDevice:
    def __init__(self, name = ''):
        print('Creating an Electronic Device')
        self.name = name

    def restart(self):
        print(f'Restarting {self.name}')

    def update_firmware(self):
        print(f'Updating firmware for {self.name}')
        ElectronicDevice.restart(self)

    
class Computer(ElectronicDevice):
    def __init__(self, name, ram):
        print('Creating a Computer')
        super().__init__(name)
        self.ram = ram
        
    def restart(self):
        print(f'Saving files on computer {self.name}')
        super().restart()

class Camera(ElectronicDevice):
    def __init__(self, name, width, height):
        print('Creating a Camera')
        super().__init__(name)
        self.width = width
        self.height = height

    def restart(self):
        print(f'Backing up photos on {self.name}')
        super().restart()
```

```python
class ElectronicDevice:
    def __init__(self, name = '', **kwargs):
        print(f'Creating an Electronic Device {kwargs=}')
        self.name = name

    def restart(self):
        print(f'Restarting {self.name}')

    def update_firmware(self):
        print(f'Updating firmware for {self.name}')
        ElectronicDevice.restart(self)

    
class Computer(ElectronicDevice):
    def __init__(self, ram, **kwargs):
        print(f'Creating a Computer {kwargs=}')
        super().__init__(**kwargs)
        self.ram = ram
        
    def restart(self):
        print(f'Saving files on computer {self.name}')
        super().restart()

class Camera(ElectronicDevice):
    def __init__(self, width, height, **kwargs):
        print(f'Creating a Camera {kwargs=}')
        super().__init__(**kwargs)
        self.width = width
        self.height = height

    def restart(self):
        print(f'Backing up photos on {self.name}')
        super().restart()


class SmartPhone(Computer, Camera):
    def __init__(self, name, ram, width, height):
        print('Creating a SmartPhone')
        args = {'name': name, 'ram': ram, 'width': width, 'height': height}
        super().__init__(**args)
```

```python
myphone = SmartPhone("Bob's smartphone", 64, 1024, 768)
```

#### Szybka dygresja - funkcje z nieskończoną ilością argumentów

```python
def some_function(x, y=1):
    return x + y
```

```python
some_function(1)
```

```python
some_function(1, y=2)
```

```python
args = [1, 2]
```

```python
some_function(args[0], args[1])
```

```python
some_function(*args) # rozpakowuje listę args i przekazuje jej elementy jako osobne argumenty do funkcji some_function
```

```python
args = {'x': 1, 'y': 2}
```

```python
some_function(**args) # rozpakowuje słownik args i przekazuje jego wartości jako osobne argumenty do funkcji some_function, używając kluczy jako nazw argumentów
```

```python
args = [1]
kwargs = {'y': 2}

some_function(*args, **kwargs)
```

```python
from abc import ABC, abstractmethod # importujemy ABC i abstractmethod, które pozwalają nam tworzyć klasy abstrakcyjne i metody abstrakcyjne

class ElectronicDevice(ABC):
    def __init__(self, name = '', **kwargs):
        print(f'Creating an Electronic Device {kwargs=}')
        self.name = name

    @abstractmethod # wymusza, aby każda klasa dziedzicząca nadpisała metodę restart (tworzy klasę abstrakcyjną, której nie można instancjonować bez nadpisania metody restart)
    def restart(self):
        print(f'Restarting {self.name}')

    def update_firmware(self):
        print(f'Updating firmware for {self.name}')
        ElectronicDevice.restart(self)

    
class Computer(ElectronicDevice):
    def __init__(self, ram, **kwargs):
        print(f'Creating a Computer {kwargs=}')
        super().__init__(**kwargs)
        self.ram = ram
        
    def restart(self):
        print(f'Saving files on computer {self.name}')
        super().restart()

class Camera(ElectronicDevice):
    def __init__(self, width, height, **kwargs):
        print(f'Creating a Camera {kwargs=}')
        super().__init__(**kwargs)
        self.width = width
        self.height = height

    def restart(self):
        print(f'Backing up photos on {self.name}')
        super().restart()


class SmartPhone(Computer, Camera):
    def __init__(self, name, ram, width, height):
        print('Creating a SmartPhone')
        args = {'name': name, 'ram': ram, 'width': width, 'height': height}
        super().__init__(**args)
```
