```python
from typing import NamedTuple
```

```python
class Student(NamedTuple):
    name: str
    age: int
    email: str
    classes: list[str]
```

```python
bob = Student(name='Bob', 
              age=21, 
              email='bob@bulszit.pl', 
              classes=['Math', 'Physics'])
```

```python
from dataclasses import dataclass, field
```

```python
@dataclass
class Student:
    name: str
    age: int
    email: str
```

```python
bob = Student(name='Bob', age=22, email='bob@bulszit.pl')
```

```python
bob.name
```

```python
bob.age = 60
```

```python
bob.age
```

```python
@dataclass(slots=True)
class Student:
    name: str
    age: int
    email: str
```

```python
bob = Student(name='Bob', age=21, email='bob@bulszit.pl')
```

```python
bob.emil = 'bob@szitbul.pl'
```

```python
@dataclass
class Student:
    name: str
    age: int = field(default_factory=lambda: 20) #nie rób tego w domu
    email: str = ''
    #classes: list[str] 
```
